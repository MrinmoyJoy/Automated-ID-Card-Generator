```python
#function to fit the image into a circle
def pic_circle_crop(base_img, im_loc):
    img=Image.open(im_loc).convert("RGB")
    npImage=np.array(img)
    h,w=img.size

    # Create same size alpha layer with circle
    alpha = Image.new('L', img.size,0)
    draw = ImageDraw.Draw(alpha)
    draw.pieslice([0,0,h,w],0,360,fill=255)

    # Convert alpha Image to numpy array
    npAlpha=np.array(alpha)

    # Add alpha layer to RGB
    npImage=np.dstack((npImage,npAlpha))
    
    photo = Image.fromarray(npImage)
    photo = photo.resize((1194, 1194))##calib

    id_card = base_img

    back_im = id_card.copy()
    back_im.paste(photo, (993, 490), photo)##calib
    return back_im
```
